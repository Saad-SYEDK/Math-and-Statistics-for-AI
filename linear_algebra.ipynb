# Linear Algebra

Linear algebra is a fundamental branch of mathematics focusing on linear equations, vector spaces, matrices, and linear transformations, is the foundational mathematical language for Artificial Intelligence (AI) and Machine Learning (ML), providing the tools to represent, manipulate, and optimize data.

It allows complex data—such as images, text, and datasets—to be structured as scalars, vectors, matrices, and tensors. These structures enable algorithms to perform efficient computations, such as training neural networks, reducing dimensionality, and solving optimization problems

## Vectors

A vector is a object having both magnitude(length) and direction.

In the field of Computer Science, a vector is a data structure (basically a array or list) or ordered list of numbers, used to represent a data point, feature, or state.

eg: consider house prices data, and we have two features, size of the hose and price, we will store a column vector, where first row will be size, and the second row will be price. We set this as an standard thorughout our project when can make it easy for processing the information.


\begin{bmatrix} w_1 \\ \vdots \\ w_n \end{bmatrix} \; or  \begin{bmatrix} Size \\ \\ Price \end{bmatrix} \;

### Vector Addition


\begin{bmatrix} x1 \\ y1 \end{bmatrix} \; +  \begin{bmatrix} x2 \\ y2 \end{bmatrix} \; = \begin{bmatrix} x1+x2 \\ y1+y2 \end{bmatrix} \;

### Vector Multiplication

We might come across multiplying a vector $\vec{V}$ by a number. eg: $2\times\vec{V}$ or `(1/3)V`. This can also be called as **Scaling** as we are simply upscaling or downscaling a vector. Here the number (2 or (1/3) in our example) is called as **Scalar**


$2 \cdot \begin{bmatrix} x \\ y \end{bmatrix} \; =  \begin{bmatrix} 2x \\ 2y \end{bmatrix} \;$

*Here, 2 is the scalar*
- Multiplying a vector by a scalar, means multiplying each components of the vector by the scalar.


### Matrix Multiplication

Multiplying multiple vectors,  also known as matrix multiplications : here's how its done

$\begin{bmatrix} a & b \\ c & d \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} =  x \begin{bmatrix} a \\ c \end{bmatrix} + y\begin{bmatrix} b \\ d \end{bmatrix} = \begin{bmatrix} ax & by \\ cx & dy \end{bmatrix} \; $

## Determinant

A determinant is a scalar value calculated from a square matrix (same number of rows and columns) that provides key information about the matrix, such as its invertibility, scaling factor, and volume transformation. Denoted as or, it is non-zero if a matrix is invertible (non-singular) and zero if singular
 
eg: for a 2x2 matrix the formula is to cross multiply and substract 

$ \begin{bmatrix} a &  b  \\  c & d \end{bmatrix} = ad - bc$


## Geometric Perspective of Determinants


### Visualizaiton: Measuring the distortion of space : The "Area Factor"

Imagine you have a simple $1 \times 1$ square on a piece of grid paper. Its area is 1.Now, apply a matrix transformation to that grid. Your square might become a tall rectangle, a tilted diamond (parallelogram), or a giant distorted shape.
- The Determinant is the new Area. * If the determinant is 3, the matrix has tripled the area of every shape in that space.
- If the determinant is 0.5, the matrix has squashed the space to half its original size.

### The Logic of Specific Values

The number you get from a determinant tells a story about what happened to the "fabric" of that space:

- Determinant = 1: The space changed shape (maybe it tilted), but the total volume/area stayed exactly the same.
- Determinant = 0 (The "Collapse"): This is the most important one. It means the matrix squashed the entire 2D world into a 1D line, or a 3D world into a 2D flat plane.
    - When space collapses, you lose information. You can’t "undo" a collapse, which is why matrices with a determinant of 0 don't have an "inverse.
- "Negative Determinant: This means the space was flipped over. Imagine turning a piece of paper over to the other  side. The "Right-hand rule" becomes a "Left-hand rule."

### Scaling to Higher Dimensions

The logic stays the same as you go up in dimensions:
- In 2D: It is the change in Area.
- In 3D: It is the change in Volume.
- In ND: It is the change in "Hyper-volume." Even if we can't visualize a 10D shape, if the determinant of a $10 \times 10$ matrix is 2, we know that whatever "10-dimensional blob" we started with is now twice as large.

### Real-Life Essence:  Where is it used?

- Changing Variables (Calculus/Physics): When scientists switch from measuring things in a "Square Grid" (like $x, y$) to a "Circular Grid" (Polar coordinates), the space gets distorted. They use the determinant (called a Jacobian) to calculate how much the "area" changed so their physics equations stay accurate.
- Computer Graphics: If a game engine wants to know if a transformation will make an object inside-out or flat, it checks the determinant.
- Detecting Errors: In complex data systems, if the determinant is 0, the system knows it has hit a "dead end" where data has been crushed and information is lost.

## Dot Products

The dot product (or scalar product) of two vectors $a=\begin{bmatrix}a_1 \ a_2,\ \dots a_n\end{bmatrix}$ and $b=\begin{bmatrix}b_1 \ b_2,\ \dots b_n\end{bmatrix}$ is a single scalar value calculated by summing the products of their corresponding components: $a \cdot b=\begin{bmatrix}a_1 b_1 \ a_2 b_2,\ \dots a_n b_n\end{bmatrix}$.

eg : 

$ \begin{bmatrix}2 \\ 7 \\ 1\end{bmatrix} \cdot \begin{bmatrix} 8\\ 2 \\ 8 \end{bmatrix} = 2 \cdot 8 + 7 \cdot 2 + 1 \cdot 8 = 38$

***Dot product of the two vectors(RESULT) can be 1. positive, 2. zero or 3. negative.***

## Geometric Prespective of Dot Product

### Dot Product: The "Shadow"

Imagine you have two vectors, A and B, starting from the same point. The dot product asks: "How much of Vector A is pointing in the same direction as Vector B?

- The Flashlight Analogy: Imagine Vector B is the floor. If you shine a flashlight straight down onto Vector B, Vector A will cast a shadow on it.
- The Calculation: The dot product is simply the length of that shadow multiplied by the length of Vector B.

### The Three Results (The Logic)

The dot product gives you a single number (a scalar), and that number tells you everything about the relationship between two directions:

- Positive Number: The vectors are "friends." They are generally pointing in the same direction (an acute angle).
- Zero: The vectors are "strangers." They are perfectly perpendicular ($90^\circ$). One has zero influence on the    other. A movement East has zero "North-ness" in it.
- Negative Number: The vectors are "enemies." They are pointing in opposite directions (an obtuse angle).

### Real-Life Essence: Where is it used?

1.  Video Games (The "Vision" Test)

In a stealth game, how does the computer know if an enemy "sees" you?
- Vector A: The direction the enemy is facing.
- Vector B: The direction from the enemy to you.
- The game calculates the dot product. If it’s high and positive, you are in their field of view. If it’s zero or negative, you are behind them or off to the side, and you're safe.

2. Data Science (The "Common Interests" Test)

Imagine Netflix represents your movie tastes as a vector in 10,000D space.
- Your Vector: [Action: 10, Romance: 2, Sci-Fi: 8]
- Movie Vector: [Action: 9, Romance: 1, Sci-Fi: 10]By taking the dot product of your vector and the movie’s vector, Netflix gets a score. The higher the score, the more "aligned" the movie is with your taste.



### Why the Formula Works

The reason we multiply the $x$-coordinates together and the $y$-coordinates together ($a_x \times b_x + a_y \times b_y$) is that we are breaking the vectors down into their simplest components to see how much they "agree" on each axis.
- If Vector A moves 5 units Right and Vector B moves 0 units Right, their agreement on the "Right-ness" is $5 \times 0 = 0$.
- The dot product sums up all these little agreements into one final score of Total Alignment.

## Cross Product

Cross product of two 3D vectors $\mathbf{a}$ and $\mathbf{b}$ is a vector $\mathbf{c}$ that is perpendicular (at a 90° angle) to both.

Formula:

 if  $\mathbf{a} = [a_1, a_2, a_3]$ and $\mathbf{b} = [b_1, b_2, b_3]$, the cross product is calculated as:$\mathbf{a} \times \mathbf{b} = [a_2b_3 - a_3b_2, \space a_3b_1 - a_1b_3, \space a_1b_2 - a_2b_1]$

## Geometric perspective of cross product

### Cross product: The "Perpendicular Creator"

The cross product is the mathematical way of saying, "Give me the direction that is perfectly unique to these two."

- The Area Connection: Just like the determinant, the length of the cross product vector is exactly equal to the area of the parallelogram formed by the two original vectors.

- The "Right-Hand Rule": If you curl the fingers of your right hand from Vector A to Vector B, your thumb points in the direction of the cross product. It is a built-in "compass" for 3D space.

### The Logic: Why does it exist?
Mathematicians "invented" this because the universe functions on rotational forces.

- Leverage (Torque): If you use a wrench to turn a bolt, you push sideways (Vector A), but the bolt moves inward (the Cross Product). The cross product calculates exactly how much "turning force" you are generating.

- Sailing/Flying: To keep a plane stable, you need to know the "Normal" vector—the line pointing straight off the wing. The cross product of the wing's length and width gives you that invisible "up" arrow.

### Real-Life Essence: Where is it used?
- Physics (Magnetism): If an electron moves through a magnetic field, the force it feels isn't in the direction it's moving—it's pushed sideways. This "Lorentz Force" is a pure cross product. It’s why electric motors spin.

- Video Games (Lighting): To make a 3D character look realistic, the computer needs to know how light bounces off its "skin." The computer takes two edges of a polygon, calculates the cross product to find the "Normal" (the direction the surface is facing), and then calculates how the light hits that direction.

- Collision Detection: It is used to determine if a 3D object is intersecting with another by checking the "perpendicular distance" between edges.

## EigenValues and EigenVectors

Eigenvalues and eigenvectors are fundamental concepts in linear algebra, used in various applications such as matrix diagonalization, stability analysis, and data analysis (e.g., Principal Component Analysis). They are associated with a square matrix and provide insights into its properties.


**Eigenvalues $(\lambda)$** are unique scalar values linked to a matrix or linear transformation. They indicate how much an eigenvector gets stretched or compressed during the transformation. The eigenvector's direction remains unchanged unless the eigenvalue is negative, in which case the direction is simply reversed.


**Eigenvectors $(v)$** are non-zero vectors that, when multiplied by a matrix, only stretch or shrink without changing direction.
- The eigenvalue must be found first before the eigenvector.

$ A\vec{x} = \lambda \vec{x}$

Where,
A -> NxN matrix or Sqaure matrix
$\vec{x}$ -> EigenVector
$\lambda$ -> EigenValue, which is basically a scalar multiplier 

## Finding EigenValue and EigenVector

eg :

Find the eigenvalues ($\lambda$) and eigenvectors ($v$) for the matrix 
$A = \begin{bmatrix} 4 & 1 \\ 2 & 3 \end{bmatrix} $


### Step 1: The Characteristic Equation

To find the eigenvalues, we must solve the equation:
$\det(A - \lambda I) = 0$ 

What is $\lambda I$?
The term $I$ is the Identity Matrix (think of it as the number "1" in matrix form). We multiply it by $\lambda$:
$\lambda I = \lambda \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} = \begin{bmatrix} \lambda & 0 \\ 0 & \lambda \end{bmatrix}$
What is $A - \lambda I$?
We subtract the $\lambda$ matrix from our original matrix $A$:

$A - \lambda I = \begin{bmatrix} 4 & 1 \\ 2 & 3 \end{bmatrix} - \begin{bmatrix} \lambda & 0 \\ 0 & \lambda \end{bmatrix} = \begin{bmatrix} 4 - \lambda & 1 \\ 2 & 3 - \lambda \end{bmatrix}$ 

Simplified rule: Just subtract $\lambda$ from the diagonal numbers.



### Step 2: Find the Eigenvalues ($\lambda$)
We calculate the determinant ($\det$) of the new matrix and set it to zero.
$\det \begin{bmatrix} 4 - \lambda & 1 \\ 2 & 3 - \lambda \end{bmatrix} = 0$ 
The calculation: $(ad - bc)$
$(4 - \lambda)(3 - \lambda) - (1)(2) = 0$ 
Expand the brackets (FOIL):
$12 - 4\lambda - 3\lambda + \lambda^2 - 2 = 0$ 
Simplify the quadratic equation:
$\lambda^2 - 7\lambda + 10 = 0$ 
Factor the quadratic:
We need two numbers that multiply to $+10$ and add to $-7$. Those are $-2$ and $-5$.
$(\lambda - 2)(\lambda - 5) = 0$ 
Solutions:
$\lambda_1 = 2, \quad \lambda_2 = 5$ 


### Step 3: Find the Eigenvector for $\lambda = 5$

We plug $\lambda = 5$ back into our $(A - \lambda I)$ matrix and solve for a vector $\begin{bmatrix} x \\ y \end{bmatrix}$ that equals zero.
$\begin{bmatrix} 4 - 5 & 1 \\ 2 & 3 - 5 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$ $\begin{bmatrix} -1 & 1 \\ 2 & -2 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$ 
Simplifying the equations:

   1. $-1x + 1y = 0 \implies x = y$
   2. $2x - 2y = 0 \implies 2x = 2y \implies x = y$

Since both equations tell us $x = y$, we can pick the simplest non-zero number. Let $x = 1$, then $y = 1$.
Eigenvector $v_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$



### Step 4: Find the Eigenvector for $\lambda = 2$

Now we plug $\lambda = 2$ into the $(A - \lambda I)$ matrix:
$\begin{bmatrix} 4 - 2 & 1 \\ 2 & 3 - 2 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$ $\begin{bmatrix} 2 & 1 \\ 2 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$ 
Simplifying the equations:
Both rows give us the same rule:
$2x + 1y = 0$ $y = -2x$ 
Choosing the values:
To make it simple, let $x = 1$. Then $y = -2(1) = -2$.
Eigenvector $v_2 = \begin{bmatrix} 1 \\ -2 \end{bmatrix}$



### Final Results

For the matrix $A = \begin{bmatrix} 4 & 1 \\ 2 & 3 \end{bmatrix}$:

* Eigenvalue $\lambda = 5$ has Eigenvector $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$
* Eigenvalue $\lambda = 2$ has Eigenvector $\begin{bmatrix} 1 \\ -2 \end{bmatrix}$
