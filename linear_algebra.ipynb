# Linear Algebra

Linear algebra is a fundamental branch of mathematics focusing on linear equations, vector spaces, matrices, and linear transformations, is the foundational mathematical language for Artificial Intelligence (AI) and Machine Learning (ML), providing the tools to represent, manipulate, and optimize data.

It allows complex data—such as images, text, and datasets—to be structured as scalars, vectors, matrices, and tensors. These structures enable algorithms to perform efficient computations, such as training neural networks, reducing dimensionality, and solving optimization problems

## Vectors

A vector is a object having both magnitude(length) and direction.

In the field of Computer Science, a vector is a data structure (basically a array or list) or ordered list of numbers, used to represent a data point, feature, or state.

eg: consider house prices data, and we have two features, size of the hose and price, we will store a column vector, where first row will be size, and the second row will be price. We set this as an standard thorughout our project when can make it easy for processing the information.


\begin{bmatrix} w_1 \\ \vdots \\ w_n \end{bmatrix} \; or  \begin{bmatrix} Size \\ Price \end{bmatrix} \;

### Vector Addition


\begin{bmatrix} x1 \\ y1 \end{bmatrix} \; +  \begin{bmatrix} x2 \\ y2 \end{bmatrix} \; = \begin{bmatrix} x1+x2 \\ y1+y2 \end{bmatrix} \;

### Vector Multiplication

We might come across multiplying a vector $\vec{V}$ by a number. eg: $2\times\vec{V}$ or `(1/3)V`. This can also be called as **Scaling** as we are simply upscaling or downscaling a vector. Here the number (2 or (1/3) in our example) is called as **Scalar**


$$2 \cdot \begin{bmatrix} x \\ y \end{bmatrix} \; =  \begin{bmatrix} 2x \\ 2y \end{bmatrix} \;$$

*Here, 2 is the scalar*
- Multiplying a vector by a scalar, means multiplying each components of the vector by the scalar.


### Matrix Multiplication

Multiplying multiple vectors,  also known as matrix multiplications : here's how its done

$$\begin{bmatrix} a \ b \\ c \ d \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} =  x \begin{bmatrix} a \\ c \end{bmatrix} + y\begin{bmatrix} b \\ d \end{bmatrix} = \begin{bmatrix} ax \ by \\ cx \ dy \end{bmatrix} \; $$

## The Geometric Perspective of - vectors and matrices:

Don't think of vectors and matrices as math problems, instead think about them like mentioned


###  Vector: A "Step" in Space

Forget about a single point on a map  A vector is a movement.
- Imagine you are standing in a perfectly empty field.
- A vector is an instruction: "Walk 3 paces East and 2 paces North".
- In 2D: You only have two directions (Left/Right, Up/Down). 
- In 3D: You add a third direction (Up/Down in the air).
- In nD: You add directions we can't see, like "Time" or "Temperature".
- A vector is just a "state" or a "direction" at a specific moment.

 ###  Matrix: The "Action"
 
 If a vector is a noun (the thing being moved), a matrix is the verb (the action of moving it). A matrix is a set of instructions that tells every possible vector in a space where to go.
 - The Grid: Imagine a rubber sheet with a grid drawn on it.
 - The Transformation: A matrix is like someone grabbing the corners of that rubber sheet and stretching, twisting, or flipping it.
 - The Result: Every point on that sheet moves to a new spot. If you know where the basic "1-unit" steps (the basis vectors) land, you know where everything lands. That is what the numbers inside a matrix represent.

### Conversion of Dimensions: "Squishing" and "Lifting"

This is where your intuition about $3 \times 2$ or $2 \times 3$ matrices becomes powerful. Dimensions aren't just places; they are levels of information.
- $2 \times 3$ (3D to 2D): This is a Shadow.
    - Think of a 3D bird flying in the sky. The sun shines down and creates a 2D shadow on the flat ground.
    - You have "lost" information (you don't know how high the bird is anymore), but you’ve simplified the data. This is how 3D video games are "projected" onto your flat TV screen.
- $3 \times 2$ (2D to 3D): This is a Sheet in Space.
    - Imagine taking a flat piece of paper (2D) and angling it so it sits inside a 3D room.
    - You are "embedding" a simple world into a more complex one.

### The $N \times N$ Dimension: The "Control Room"

When we get to $10 \times 10$ or $100 \times 100$, we stop trying to draw it and start using it for Complexity.<br>

## Dot Products

The dot product (or scalar product) of two vectors $a=\begin{bmatrix}a_1 \ a_2,\ \dots a_n\end{bmatrix}$ and $b=\begin{bmatrix}b_1 \ b_2,\ \dots b_n\end{bmatrix}$ is a single scalar value calculated by summing the products of their corresponding components: $a \cdot b=\begin{bmatrix}a_1 b_1 \ a_2 b_2,\ \dots a_n b_n\end{bmatrix}$.

eg : 

$ \begin{bmatrix}2 \\ 7 \\ 1\end{bmatrix} \cdot \begin{bmatrix} 8\\ 2 \\ 8 \end{bmatrix} = 2 \cdot 8 + 7 \cdot 2 + 1 \cdot 8 = 38$

***Dot product of the two vectors(RESULT) can be 1. positive, 2. zero or 3. negative.***

## Geometric Prespective of Dot Product

### Dot Product: The "Shadow"

Imagine you have two vectors, A and B, starting from the same point. The dot product asks: "How much of Vector A is pointing in the same direction as Vector B?

- The Flashlight Analogy: Imagine Vector B is the floor. If you shine a flashlight straight down onto Vector B, Vector A will cast a shadow on it.
- The Calculation: The dot product is simply the length of that shadow multiplied by the length of Vector B.

### The Three Results (The Logic)

The dot product gives you a single number (a scalar), and that number tells you everything about the relationship between two directions:

- Positive Number: The vectors are "friends." They are generally pointing in the same direction (an acute angle).
- Zero: The vectors are "strangers." They are perfectly perpendicular ($90^\circ$). One has zero influence on the    other. A movement East has zero "North-ness" in it.
- Negative Number: The vectors are "enemies." They are pointing in opposite directions (an obtuse angle).

### Real-Life Essence: Where is it used?

1.  Video Games (The "Vision" Test)

In a stealth game, how does the computer know if an enemy "sees" you?
- Vector A: The direction the enemy is facing.
- Vector B: The direction from the enemy to you.
- The game calculates the dot product. If it’s high and positive, you are in their field of view. If it’s zero or negative, you are behind them or off to the side, and you're safe.

2. Data Science (The "Common Interests" Test)

Imagine Netflix represents your movie tastes as a vector in 10,000D space.
- Your Vector: [Action: 10, Romance: 2, Sci-Fi: 8]
- Movie Vector: [Action: 9, Romance: 1, Sci-Fi: 10]By taking the dot product of your vector and the movie’s vector, Netflix gets a score. The higher the score, the more "aligned" the movie is with your taste.



### Why the Formula Works

The reason we multiply the $x$-coordinates together and the $y$-coordinates together ($a_x \times b_x + a_y \times b_y$) is that we are breaking the vectors down into their simplest components to see how much they "agree" on each axis.
- If Vector A moves 5 units Right and Vector B moves 0 units Right, their agreement on the "Right-ness" is $5 \times 0 = 0$.
- The dot product sums up all these little agreements into one final score of Total Alignment.
