## Types of Data

We can split based on the nature of data, it can be of two types:
1. Qualitaive (or Categorical)<br>
   Based on **level of measurement**, This can further be classified into two:
    1. Nominal - They do not have inherent rank, eg:  Blood group(O, B, A), eye color(brown, blue), etc
    2. Ordinal - They have natural logical order, eg: school grades(A, D), movie rating(8, 6), etc
2. Quantitative (or Numerical)
