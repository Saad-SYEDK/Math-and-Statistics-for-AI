## Types of Data

We can split based on the nature of data, are we **measuring** them or **categorizing**? it can be of two types:
1. Qualitaive (or Categorical)<br>
   This is non-numerical data, we cannot perform meaningful math on it, Based on **level of measurement**, This can further be classified into two:
    1. Nominal (Labelled) - They do not have inherent rank, eg:  Blood group(O, B, A), eye color(brown, blue), etc
    2. Ordinal (Ranked)- They have natural logical order, eg: school grades(A, D), movie rating(8, 6), etc
2. Quantitative (or Numerical)<br>
   Numerical data on which we can perfom mathematical operations on, This can further be classified into two:
    1. Discrete - Contains distinct separate values with gaps between them. Countable whole numbers. eg : no of cars
    2. Continious - Contains unbroken sequence of values within a range eg: Heights of person

    *NOTE*:<br>
    1. Height is Continuous because...
        Between any two heights (e.g., 170 cm and 171 cm), there are infinite possible values. You can be 170.5 cm, 170.55 cm, or 170.552 cm. The only limit is the precision of your measuring tool.
    2. Number of Cars is Discrete because...
        It is based on counting units. You can have 1 car or 2 cars, but it is mathematically impossible to have 1.5 cars. Even if you have a million cars, you are still moving in distinct "jumps" of +1. There is no value that exists between the integers 1 and 2.
