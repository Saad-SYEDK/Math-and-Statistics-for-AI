# Calculus

Calculus is the mathematical study of continuous change, focusing on rates of change (differential calculus) and the accumulation of quantities (integral calculus).

***Calculus is the Mathematics of Change. If Linear Algebra is about stretching and warping space, Calculus is about zooming in until the curves look like straight lines.***

## Derivatives

Derivatives in Maths refers to the instantaneous rate of change of a quantity with respect to the other. It helps to investigate the moment by moment nature of an amount.



### Derivatives Formulas


$\frac{d}{dx}(k) = 0$, where $k$ is any constant
$\frac{d}{dx}(x) = 1$
$\frac{d}{dx}(x^n) = nx^{n-1}$
$\frac{d}{dx}(kx) = k$, where $k$ is any constant
$\frac{d}{dx}(\sqrt{x}) = \frac{1}{2\sqrt{x}}$
$\frac{d}{dx}\left(\frac{1}{x}\right) = -\frac{1}{x^2}$
$\frac{d}{dx}(\ln x) = \frac{1}{x}, x > 0$
$\frac{d}{dx}(e^x) = e^x$
$\frac{d}{dx}(a^x) = a^x \ln a$



### Derivatives of Trigonometric Functions




$\frac{d}{dx} (\sin x) = \cos x$
$\frac{d}{dx} (\cos x) = -\sin x$
$\frac{d}{dx} (\tan x) = \sec^2 x$
$\frac{d}{dx} (\csc x) = -\csc x \cot x$
$\frac{d}{dx} (\sec x) = \sec x \tan x$
$\frac{d}{dx} (\cot x) = -\csc^2 x$



## Integrals

Integrals represents the accumulation of quantities, such as areas under curves or net changes, and act as the inverse operation to differentiation (antiderivatives)
