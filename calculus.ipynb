# Calculus

Calculus is the mathematical study of continuous change, focusing on rates of change (differential calculus) and the accumulation of quantities (integral calculus).

***Calculus is the Mathematics of Change. If Linear Algebra is about stretching and warping space, Calculus is about zooming in until the curves look like straight lines.***

In this notes we will only make point notes for recalling purpose, this notes is not intended to explain calculus in detail.

You may use this as a reference for which topics to study.

## Derivatives

Derivatives in Maths refers to the instantaneous rate of change of a quantity with respect to the other. It helps to investigate the moment by moment nature of an amount.



### Derivatives Formulas

**Power Rule:** $\frac{d}{dx} (x^n) = nx^{n-1}$
