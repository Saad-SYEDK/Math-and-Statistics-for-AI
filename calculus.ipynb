# Calculus

Calculus is the mathematical study of continuous change, focusing on rates of change (differential calculus) and the accumulation of quantities (integral calculus).

***Calculus is the Mathematics of Change. If Linear Algebra is about stretching and warping space, Calculus is about zooming in until the curves look like straight lines.***

In this notes we will only make point notes for recalling purpose, this notes is not intended to explain calculus in detail.

You may use this as a reference for which topics to study.

## Derivatives

Derivatives in Maths refers to the instantaneous rate of change of a quantity with respect to the other. It helps to investigate the moment by moment nature of an amount.



### Derivatives Formulas


$\frac{d}{dx}(k) = 0$, where $k$ is any constant
$\frac{d}{dx}(x) = 1$
$\frac{d}{dx}(x^n) = nx^{n-1}$
$\frac{d}{dx}(kx) = k$, where $k$ is any constant
$\frac{d}{dx}(\sqrt{x}) = \frac{1}{2\sqrt{x}}$
$\frac{d}{dx}\left(\frac{1}{x}\right) = -\frac{1}{x^2}$
$\frac{d}{dx}(\ln x) = \frac{1}{x}, x > 0$
$\frac{d}{dx}(e^x) = e^x$
$\frac{d}{dx}(a^x) = a^x \ln a$



### Derivatives of Trigonometric Functions




$\frac{d}{dx} (\sin x) = \cos x$
$\frac{d}{dx} (\cos x) = -\sin x$
$\frac{d}{dx} (\tan x) = \sec^2 x$
$\frac{d}{dx} (\csc x) = -\csc x \cot x$
$\frac{d}{dx} (\sec x) = \sec x \tan x$
$\frac{d}{dx} (\cot x) = -\csc^2 x$



## Integrals

Integrals represents the accumulation of quantities, such as areas under curves or net changes, and act as the inverse operation to differentiation (antiderivatives)

### Integral Formulae

Similar to differentiation formulas, we have integral formulas 

$\int a \, dx = ax + C$
$\int \frac{1}{x} \, dx = \ln |x| + C$
$\int e^x \, dx = e^x + C$
$\int a^x \, dx = \frac{a^x}{\ln a} + C$
$\int \ln x \, dx = x \ln x - x + C$
$\int \sin x \, dx = -\cos x + C$
$\int \cos x \, dx = \sin x + C$
$\int \tan dx = \ln |\sec x| + C \text{ or } -\ln |\cos x| + C$
$\int \cot x \, dx = \ln |\sin x| + C$
$\int \sec x \, dx = \ln |\sec x + \tan x| + C$
$\int \csc x \, dx = \ln |\csc x - \cot x| + C$
$\int \sec^2 x \, dx = \tan x + C$
$\int \sec x \tan x \, dx = \sec x + C$
$\int \csc^2 x \, dx = -\cot x + C$
$\int \tan^2 x \, dx = \tan x - x + C$
$\int \frac{dx}{\sqrt{a^2 - x^2}} = \arcsin\left(\frac{x}{a}\right) + C$
$\int \frac{dx}{a^2 + x^2} = \frac{1}{a} \arctan\left(\frac{x}{a}\right) + C$

## A Better perspective

### Derivatives

To understand why the derivative of $x^3$ is $3x^2$, Imagine you have a physical cube with a side length of $x$. Its volume is $x^3$.

Now, imagine you increase the size of that cube by a tiny, microscopic amount, which we call $dx$ (a "tiny nudge" in $x$).

When the cube grows, where does the new volume go? It gets added to the faces:

1. A thin layer is added to the Top ($x \times x \times dx$).

2. A thin layer is added to the Front ($x \times x \times dx$).

3. A thin layer is added to the Side ($x \times x \times dx$).

Since there are 3 faces getting a new layer of area $x^2$, the total increase in volume is roughly:

$$d(\text{Volume}) \approx 3x^2 \cdot dx$$

If you divide both sides by that tiny nudge $dx$, you get the rate of change:

$$\frac{dy}{dx} = 3x^2$$

(There are tiny slivers on the edges and corners too, but as $dx$ shrinks toward zero, those slivers become so small they basically vanish. Only the 3 big faces matter.)

**In other words**
A Derivative ($dy/dx$) is simply a Speedometer.

- The Function ($x^2, x^3$): Tells you where you are (Distance).
- The Derivative ($2x, 3x^2$): Tells you how fast you are moving at that exact split-second (Velocity).

The "Logic" was invented because the universe doesn't move in chunks; it moves smoothly. If you throw a ball, its speed changes every microsecond. Calculus allows us to freeze time and measure that change "at an instant."

### Intergrals - The reverse

If the Derivative is about breaking things down into tiny slices to see how fast they change, the Integral ($\int$) is about gluing them back together to find the total.

**Derivative**: "I have a volume; how fast is it growing?"

**Integral**: "I know how fast it's growing; what is the total volume I've built?"

Think of an Integral as a Salami Slicer. If you know the area of every tiny circular slice of a salami, and you add them all up from the start to the end, you get the total volume of the meat.
