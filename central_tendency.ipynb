# Central Tendency

Central Tendency measures and identifies the center or typical value of a dataset, summarizing it with a single value.

## Mean

Mean is an average calculated by summing a set of numerical values and dividing by the total count of those values. It represents the "center" of a data set, often used to summarize data in statistics, finance, and daily life. 

Calculation: 
$ Mean = \frac{Sum~of~all~values}{Total~number~of~values}$ 

Example: For the set {2, 4, 4, 5, 5}, the sum is 20. Divided by 5, the mean is 4.


## Median

The median is the middle value in an ordered data set, separating the higher half from the lower half.

Sort the data is ascending order and the middle value is the Median, and if there are even number of values then median will be the center of the two values.

example : <br> 
odd set -> {2, 4, 4, 5, 5}, median = 4 <br>
even set -> {2, 4, 5, 5}, median = (4+5)/2 = 4.5

`*Note: It is necessary to first sort the data in ascending order or else the answer will be wrong*`
